```python
import json
import numpy as np
import scipy.io
import csv
import os

#NumConds = 2
baseline = ['A1_shareEncoder','A2_targetonly','A3_contextonly','A4_nodettach']
condition = ['IC','gravity','coocurrence','coocurrence_and_gravity','size','no_context']
#result_mean = np.zeros((len(baseline),2,len(condition)))
#result_std = np.zeros((len(baseline),2,len(condition)))

for selected in range(len(baseline)):
    print(baseline[selected])
    for cond in range(len(condition)):
        print(condition[cond])
        
        imagenamelist = []
        groundtruthlist = []
        predictedlist = []
        correctlist = []
        
        with open('phillip/ablation/' + baseline[selected] + '/virtualhome_' + condition[cond] + '_individual_scores.json') as f:
            scoredict = json.load(f)
            #print(scoredict)
        with open('phillip/id2filename/virtualhome_' + condition[cond] + '_id2filename.json') as f:
            imgdict = json.load(f)
            #print(imgdict)

        for i, annid in enumerate(imgdict.keys()):
            #print(scoredict[i])
            #print(imgdict[annid])
            imagenamelist.append(imgdict[annid]) 
            groundtruthlist.append(scoredict[i][3])
            predictedlist.append(scoredict[i][5])
            if isinstance(scoredict[i][1], (bool)):
                if scoredict[i][1]:
                    correctlist.append(1)
                else:
                    correctlist.append(0)
            else:
                correctlist.append(scoredict[i][1])    
        
        #print(correctlist)
        newfilePath = '/home/mengmi/Projects/Proj_context3/VirtualHome/matlab/Mat/Ajason_' + baseline[selected] + '_' + condition[cond] + '.csv'
        print(newfilePath)
        with open(newfilePath, "w") as f:
            fieldnames = ['imagename','groundtruth','predicted','correctness']
            #writer = csv.writer(f)
            writer = csv.DictWriter(f, fieldnames=fieldnames)
            writer.writeheader()
            for i, content in enumerate(imagenamelist):
                writer.writerow({'imagename': imagenamelist[i],
                                'groundtruth': groundtruthlist[i],
                                 'predicted': predictedlist[i],
                                 'correctness': correctlist[i]})
        
#print(result_mean)
#print(result_std)
#scipy.io.savemat('phillip/COCOstuff_trained_zip/zip_format/resultMat.mat', {'result_mean': result_mean, 'result_std': result_std})
            
```
