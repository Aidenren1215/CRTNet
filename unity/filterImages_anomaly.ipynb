```python
import pickle
import pandas
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
from shutil import copyfile
import csv
import os.path
from os import path

apartmentlist = range(0,7) #total: 7
rootdir = '/media/mengmi/KLAB15/Mengmi/Proj_context3/VirtualHome/unity/'
Jasondir = rootdir + "jason_anomaly/"
Imgdir = "stimulus_anomaly/"
classnamelist = []
roomlist = []
apartidlist = []
imagenamelist = []

leftlist = []
toplist = []
rightlist = []
bottomlist = []

startStr = "img_"
endStr = "_prefab_"

for apartmentid in apartmentlist:
    f = open(rootdir + "filtered_anomaly/apartment_" + str(apartmentid) + "_good.txt", "r")    
    for x in f:  
        x =x[len(rootdir+Imgdir):]
        infor = pickle.load( open( (Jasondir + x).rstrip('.png\n') + '.pkl', "rb" ) )        
        targetclassname = infor['target_node']['class_name']
        targetroom = infor['targetroomname']
        
        targetid = int(x[x.find(startStr)+len(startStr):x.rfind(endStr)])
        JasonData = infor['JasonData']
        for obj in JasonData.items():
            if obj[1]['prefab_id'] == targetid:            
                left = obj[1]['bbox'][2]
                top = obj[1]['bbox'][0]
                right = obj[1]['bbox'][3]
                bottom = obj[1]['bbox'][1]
                break 
                
        
        roomlist.append(targetroom)
        apartidlist.append(apartmentid)        
        classnamelist.append(targetclassname)
        imagenamelist.append(Imgdir + x)
        
        leftlist.append(left)
        toplist.append(top)
        rightlist.append(right)
        bottomlist.append(bottom)
        
        
Jasondir = rootdir + "jason_anomaly_wall/"
Imgdir = "stimulus_anomaly_wall/"

for apartmentid in apartmentlist:
    f = open(rootdir + "filtered_anomaly_wall/apartment_" + str(apartmentid) + "_good.txt", "r")    
    for x in f:  
        x =x[len(rootdir+Imgdir):]
        xname = (Jasondir + x).rstrip('.png\n') + '.pkl'
        print(xname)
#         imgxname = rootdir + Imgdir+x
#         print(imgxname)
#         if not os.path.exists(xname):
#             print('aaaaaaaaaaaaaaaaaaaaaa')
#             continue
        
#         if not os.path.exists(imgxname):
#             print('ddddddddddddddddddddd')
#             continue
            
        infor = pickle.load( open(xname, "rb" ) )        
        targetclassname = infor['target_node']['class_name']
        targetroom = infor['targetroomname']
        targetbbox = infor['target_bbox']
        #left, top, right, bottom
        left = targetbbox[0]
        top = targetbbox[1]
        right = targetbbox[2]
        bottom = targetbbox[3]                  
        
        roomlist.append(targetroom)
        apartidlist.append(apartmentid)        
        classnamelist.append(targetclassname)
        print(Imgdir+x)
        imagenamelist.append(Imgdir + x)
        
        leftlist.append(left)
        toplist.append(top)
        rightlist.append(right)
        bottomlist.append(bottom)

        
```
