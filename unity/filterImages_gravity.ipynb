```python
import pickle
import pandas
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
from shutil import copyfile
import csv

apartmentlist = range(0,7) #total: 7
rootdir = '/media/mengmi/KLAB15/Mengmi/Proj_context3/VirtualHome/unity/'
Jasondir = rootdir + "jason_gravity/"
Imgdir = "stimulus_gravity/"
classnamelist = []
roomlist = []
apartidlist = []
imagenamelist = []
surfacelist = []
leftlist = []
toplist = []
rightlist = []
bottomlist = []

startStr = "img_"
endStr = "_prefab_"

BadClassName = ['toy','mincedmeat','mousemat',
                'radio','longboard','cutleryfork','coffeepot','breadslice','banana']

for apartmentid in apartmentlist:
    f = open(rootdir + "filtered_gravity/apartment_" + str(apartmentid) + "_good.txt", "r")    
    for x in f:        
        infor = pickle.load( open( (Jasondir + x).rstrip('.png\n') + '.pkl', "rb" ) )  
        targetid = int(x[x.find(startStr)+len(startStr):x.rfind(endStr)])
        JasonData = infor['JasonData']
        for obj in JasonData.items():
            if obj[1]['prefab_id'] == targetid:            
                left = obj[1]['bbox'][2]
                top = obj[1]['bbox'][0]
                right = obj[1]['bbox'][3]
                bottom = obj[1]['bbox'][1]
                break                
            
        targetclassname = infor['target_node']['class_name']
        targetroom = infor['targetroomname']
        targetsurface = infor['dest_surf']
        if targetclassname in BadClassName:
            continue
        
        roomlist.append(targetroom)
        apartidlist.append(apartmentid)        
        classnamelist.append(targetclassname)
        imagenamelist.append(Imgdir + x)
        surfacelist.append(targetsurface)
        leftlist.append(left)
        toplist.append(top)
        rightlist.append(right)
        bottomlist.append(bottom)
 
Jasondir = rootdir + "jason_gravity_ori/"
Imgdir = "stimulus_gravity_ori/"
for apartmentid in apartmentlist:
    f = open(rootdir + "filtered_gravity_ori/apartment_" + str(apartmentid) + "_good.txt", "r")    
    for x in f:        
        infor = pickle.load( open( (Jasondir + x).rstrip('.png\n') + '.pkl', "rb" ) )        
        targetclassname = infor['target_node']['class_name']
        targetroom = infor['targetroomname']
        targetsurface = infor['dest_surf']
        targetid = int(x[x.find(startStr)+len(startStr):x.rfind(endStr)])
        JasonData = infor['JasonData']
        for obj in JasonData.items():
            if obj[1]['prefab_id'] == targetid:            
                left = obj[1]['bbox'][2]
                top = obj[1]['bbox'][0]
                right = obj[1]['bbox'][3]
                bottom = obj[1]['bbox'][1]
                break
                
        if targetclassname in BadClassName:
            continue
        
        roomlist.append(targetroom)
        apartidlist.append(apartmentid)        
        classnamelist.append(targetclassname)
        imagenamelist.append(Imgdir + x)
        surfacelist.append(targetsurface)
        leftlist.append(left)
        toplist.append(top)
        rightlist.append(right)
        bottomlist.append(bottom)
```

```python
print(len(roomlist))
rows = zip(classnamelist,apartidlist,roomlist,surfacelist,imagenamelist,leftlist,toplist,rightlist,bottomlist)
newfilePath = '/home/mengmi/Projects/Proj_context3/VirtualHome/matlab/Mat/jason_gravity_combined.csv'

with open(newfilePath, "w") as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
