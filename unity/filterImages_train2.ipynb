```python
import pickle
import pandas
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
from shutil import copyfile
import csv
import os

apartmentlist = range(0,7) #total: 7
rootdir = '/media/mengmi/KLAB15/Mengmi/Proj_context3/VirtualHome/unity/'
Jasondir = rootdir + "jason_train_2/"
Imgdir = rootdir + "stimulus_train_2/"
classnamelist = []
roomlist = []
apartidlist = []
imagenamelist = []

leftlist = []
toplist = []
rightlist = []
bottomlist = []


for apt in range(7):
    imgfolder = rootdir + 'stimulus_train_2/apartment_' + str(apt) + '/'
    jasonfolder = rootdir + 'jason_train_2/apartment_' + str(apt) + '/'
    
    for f, filename in enumerate(os.listdir(imgfolder)):
        
        print('processing apt: '+ str(apt) +'; file: '+ str(f))
        
        x = imgfolder + filename
        #print(filename)
        #print(x)
        x =x[len(Imgdir):]
        #(x)
        #print((Jasondir + x).rstrip('.png\n') + '.pkl')
        infor = pickle.load( open( (Jasondir + x).rstrip('.png\n') + '.pkl', "rb" ) )        
        targetclassname = infor['target_node']['class_name']
        targetroom = infor['targetroomname']
        targetbbox = infor['target_bbox']
              
        #left, top, right, bottom
        left = targetbbox[0]
        top = targetbbox[1]
        right = targetbbox[2]
        bottom = targetbbox[3]                
        
        roomlist.append(targetroom)
        apartidlist.append(apt)        
        classnamelist.append(targetclassname)
        imagenamelist.append(x)
        
        leftlist.append(left)
        toplist.append(top)
        rightlist.append(right)
        bottomlist.append(bottom)
        
        

 
```

```python
newfilePath = '/home/mengmi/Projects/Proj_context3/VirtualHome/matlab/Mat/jason_trained2.csv'

with open(newfilePath, "w") as f:
    fieldnames = ['classname', 'apartmentid','room','imagename','left','top','right','bottom']
    #writer = csv.writer(f)
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    #writer.writerow({'classname': classnamelist, 'apartmentid': apartidlist})
    for i, content in enumerate(classnamelist):
        writer.writerow({'classname': classnamelist[i],
                        'apartmentid': apartidlist[i],
                         'left': leftlist[i],
                         'top': toplist[i],
                         'right': rightlist[i],
                        'bottom': bottomlist[i],
                        'room': roomlist[i],
                        'imagename':imagenamelist[i]})
```
